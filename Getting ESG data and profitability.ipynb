```python
import eikon as ek
import pandas as pd
import datetime as dt
import logging
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
# Get the list of ticker symbols for the SNP500 index constituents
snp500 = ek.get_data('0#.SPX', 'TR.IndexConstituentRIC')[0].set_index('Constituent RIC')
snp500
```

```python
# Set the timeout limit to 10000 seconds
options = {'timeout': 10000}

# Get the data for each SNP500 constituent
data = pd.DataFrame()
for symbol in snp500.index:
#     print(symbol)
    try:
        result = ek.get_data(symbol, fields=[
        'TR.CommonName','TR.TRBCEconomicSector', 'TR.BetaFiveYear', 'TR.TRESGScore(Period=FY0,Frq=FY,SDate=0,EDate=-4)',
        'TR.RevenueActValue(Period=FY0,Frq=FY,SDate=0,EDate=-4,Scale=6,Curn=Native)', 'TR.F.OpProfBefNonRecurIncExpn(SDate=0,EDate=-4,Period=FY0,Frq=FY,Curn=Native,Scale=6,ReportingState=Orig)', 'TR.TRESGCScore(Frq=FY,SDate=0,EDate=-4,Period=FY0)',
        'TR.F.MktCap(Frq=FY,SDate=0,EDate=-4,Period=FY0,Scale=6)',
        'TR.TotalDebtToEBITDA(Frq=FY,SDate=0,EDate=-4)',
        'TR.F.TotCurrAssets(SDate=0,EDate=-4,Frq=FY,Period=FY0,Scale=6)',
        'TR.F.TotCurrLiab(SDate=0,EDate=-4,Frq=FY,Period=FY0,Scale=6)',
        'TR.PtoCPXMeanEst(SDate=0,EDate=-4,Period=FY0,Frq=FY)',
        'TR.F.CAPEXTot(SDate=0,EDate=-4,Scale=6,Frq=FY,Period=FY0)',
        'TR.F.IncTaxRatePct(Period=FY0,Frq=FY,SDate=0,EDate=-4)',
        'TR.RevenueMean(SDate=0,EDate=-4,Period=FY0,Frq=FY,Scale=6)',
        'TR.F.WkgCap(SDate=0,EDate=-4,Scale=6,Period=FY0,Frq=FY)'
        'TR.ROCEActValue(SDate=0,EDate=-4,Period=FY0,Frq=FY)',
        'TR.F.WkgCapPctTotRev(SDate=0,EDate=-4,Period=FY0,Frq=CY)',
        'TR.F.NetDebttoTotCap(Period=FY0,Frq=FY,SDate=0,EDate=-4)'])
        
        result_df = pd.concat(result, axis = 1)
        data= pd.concat([data, result_df], axis = 0)
        print(data)
    except Exception as e:
        print(f"Error getting data for {symbol}: {e}")
        traceback.print_exc()
        pass
    
```
